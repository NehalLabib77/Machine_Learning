Importing libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

```python
X, y = datasets.load_diabetes(return_X_y=True) # Scikit-learn’s built-in diabetes dataset.

# convert to Dataframe 
df = pd.DataFrame(X,columns=datasets.load_diabetes().feature_names)
df['target'] = y

print(df.head())
print(X.shape)
print(X[0])
```

```python
X = X[:, [2]]   # notice the double brackets
print(X.shape)  # (442, 1)


X = X.reshape((-1, 1))
print(X.shape)
```

```python
# Train - Test data 
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
# Model Train
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict 
y_pred = model.predict(X_test)
```

```python
# Plot
# Sort X_test and predictions for a clean regression line

sorted_idx = X_test[:, 0].argsort()
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test[sorted_idx], y_pred[sorted_idx], color='blue', linewidth=3)
plt.xlabel('Scaled BMI')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
