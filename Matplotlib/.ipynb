**IMPORTING LIBRARIES**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**BASIC GRAPH**

```python
plt.plot([1,2,3],[4,5,6])
```

**DIFFERENT KINDS OF TAG**

```python
# Define data for x and y axes
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Plot the data
plt.plot(x, y)

# Set the title with custom font
plt.title("Our First Graph!", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# Label the x-axis
plt.xlabel("X Axis")

# Label the y-axis
plt.ylabel("Y Axis")

# Display the plot
plt.show()
```

**TICKMARKS(SCALE OF YOUR GRAPH)**

```python
# Data (x and y are already defined in earlier cells)
# x = [0,1,2,3,4]
# y = [0,2,4,6,8]

# Tick configuration: set which tick marks appear on the x- and y-axes
plt.xticks([0,1,2,3,4])             # x-axis ticks at 0,1,2,3,4
plt.yticks([0,2,4,6,8,10])          # y-axis ticks at 0,2,4,6,8,10

# Plot styling:
# label       -> text shown in the legend for this line
# color       -> line color
# linewidth   -> thickness of the line
# marker      -> marker symbol for data points ('.' is a small dot)
# linestyle   -> line style ('--' is dashed)
# markersize  -> size of the marker symbol
# markeredgecolor -> color of the marker border
plt.plot(x, y,
         label='2x',
         color='red',
         linewidth=2,
         marker='.',
         linestyle='--',
         markersize=15,
         markeredgecolor='black')

# Title & axis labels (font customizations)
plt.title("Our First Graph!", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel("X Axis")                # label for x-axis
plt.ylabel("Y Axis")                # label for y-axis

# Legend & display
plt.legend()                        # shows a legend box with the label(s)
plt.show()                          # renders the plot inline in the notebook
```

**SHORTHAND NOTATION**

```python
# === DATA PREPARATION ===
# Primary x and y data for the linear 2x line
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Secondary x values for plotting a quadratic curve (0 to 4 step 0.5)
# Using numpy to get a denser set of x values for the parabola
x2 = np.arange(0,4.5,0.5)
print(x2)

# === FIGURE SETUP ===
# Resize the output figure (width=5in, height=3in) and set resolution (dpi)
plt.figure(figsize=(5,3), dpi=100)

# === AXIS TICKS ===
# Explicitly set which tick marks appear on each axis
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

# === PLOTTING LINES (MAIN SUBPLOTS) ===
# Plot part of the quadratic in solid red with a legend entry:
# x2[:6] selects the first 6 points (0..2.5) and x2[:6]**2 computes y = x^2
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')

# x2[4:] starts at index 4 (2.0) so the dashed segment overlaps slightly for visual continuity
plt.plot(x2[4:], x2[4:]**2, 'r--')

# 'b.--' = blue color, dot markers, and dotted line; label used in legend
plt.plot(x, y, 'b.--', label='2x')

# === STYLING & ANNOTATIONS ===
# Title and axis labels with a custom font and size for the title
plt.title("Our First Graph!", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Show a legend to identify plotted lines
plt.legend()

# === OUTPUT ===
# Save a high-resolution copy of the figure to disk
plt.savefig('mygraph.png', dpi=300)

# Render the plot inline in the notebook
plt.show()
```

## **BAR CHART**

```python
# Category labels and their corresponding heights for the bar chart
labels = ['A', 'B', 'C']
values = [1, 4, 2]

# Create the bar chart; 'bars' is a BarContainer of Rectangle artists
bars = plt.bar(labels, values)

# Hatch patterns add visual texture to the bars (one pattern per bar)
patterns = ['/', 'O', '*']

# Apply a hatch pattern to each bar by popping the next pattern
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# Save the current figure to a PNG file with high resolution
plt.savefig('barchart.png', dpi=300)

# Create a new figure with a specific size.
# Note: calling plt.figure() here creates a new empty figure;
# plt.show() will display that new empty figure. To display the saved chart
# inline, remove or move this plt.figure(...) call before plotting.
plt.figure(figsize=(5,3))

# Display the figure(s) inline in the notebook
plt.show()
```
